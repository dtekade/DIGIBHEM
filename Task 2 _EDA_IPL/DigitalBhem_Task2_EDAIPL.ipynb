```python
# EDA
```

```python
# Number of matches played each season
matches_per_season = matches['season'].value_counts().sort_index()

# Plotting the number of matches per season
plt.figure(figsize=(12, 6))
sns.countplot(data=matches, x='season', palette='viridis')
plt.title('Number of Matches Played Each Season')
plt.xlabel('Season')
plt.ylabel('Number of Matches')
plt.xticks(rotation=45)
plt.show()
```

```python
# Number of wins per team
wins_per_team = matches['winner'].value_counts()

# Plotting the number of wins per team
plt.figure(figsize=(12, 6))
sns.countplot(data=matches, x='winner', palette='viridis', order=wins_per_team.index)
plt.title('Number of Wins Across Teams')
plt.xlabel('Team')
plt.ylabel('Number of Wins')
plt.xticks(rotation=90)
plt.show()
```

```python
# Top batsmen by runs
top_batsmen = delivery.groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False).head(10)

# Plotting top batsmen
plt.figure(figsize=(12, 6))
sns.barplot(x=top_batsmen.index, y=top_batsmen.values, palette='viridis')
plt.title('Top 10 Batsmen by Runs')
plt.xlabel('Batsman')
plt.ylabel('Total Runs')
plt.xticks(rotation=45)
plt.show()
```

```python
# Top bowlers by wickets
# Note: In the deliveries dataset, dismissals are recorded in the column 'dismissal_kind', but not all dismissals are attributed to the bowler (e.g., run out).

# Filter out 'run out' and count wickets
valid_dismissals = delivery[delivery['dismissal_kind'].notnull() & (delivery['dismissal_kind'] != 'run out')]
top_bowlers = valid_dismissals['bowler'].value_counts().head(10)

# Plotting top bowlers
plt.figure(figsize=(12, 6))
sns.barplot(x=top_bowlers.index, y=top_bowlers.values, palette='viridis')
plt.title('Top 10 Bowlers by Wickets')
plt.xlabel('Bowler')
plt.ylabel('Total Wickets')
plt.xticks(rotation=45)
plt.show()
```

```python
# Factors Contributing to Team Wins or Losses:
# Win percentage by team
total_matches_per_team = matches['team1'].value_counts() + matches['team2'].value_counts()
win_percentage = (wins_per_team / total_matches_per_team) * 100
win_percentage = win_percentage.dropna().sort_values(ascending=False)

# Plotting win percentage by team
plt.figure(figsize=(12, 6))
sns.barplot(x=win_percentage.index, y=win_percentage.values, palette='viridis')
plt.title('Win Percentage by Team')
plt.xlabel('Team')
plt.ylabel('Win Percentage')
plt.xticks(rotation=90)
plt.show()


```

```python
# Toss decision impact
toss_decision = matches.groupby(['toss_decision', 'winner']).size().unstack().fillna(0)
toss_decision_percentage = (toss_decision.div(toss_decision.sum(axis=1), axis=0) * 100).round(2)

# Plotting toss decision impact
toss_decision.plot(kind='bar', stacked=True, figsize=(15, 6), colormap='viridis')
plt.title('Impact of Toss Decision on Wins')
plt.xlabel('Toss Decision')
plt.ylabel('Number of Wins')
plt.xticks(rotation=0)
plt.show()
```

```python
# Team performance over seasons
team_performance = matches.groupby(['season', 'winner']).size().unstack().fillna(0)

# Plotting team performance over seasons
plt.figure(figsize=(15, 8))
team_performance.plot(kind='bar', stacked=True, figsize=(15, 8), colormap='viridis')
plt.title('Team Performance Over Seasons')
plt.xlabel('Season')
plt.ylabel('Number of Wins')
plt.xticks(rotation=45)
plt.legend(title='Teams', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
# Merge the deliveries dataset with matches dataset to get the season information
deliveries = delivery.merge(matches[['id', 'season']], left_on='match_id', right_on='id')

```

```python
# Verify the merge and check the columns
print("Deliveries Dataset After Merge:")
print(deliveries.head())
print("\nColumns in Deliveries Dataset:")
print(deliveries.columns)

```

```python
# Check if the season column is present
if 'season' in deliveries.columns:
    print("\nSeason column is present in the deliveries dataset.")
else:
    print("\nSeason column is NOT present in the deliveries dataset.")
```

```python
# Top batsmen each season
top_batsmen_season = deliveries.groupby(['season', 'batsman'])['batsman_runs'].sum().reset_index()
top_batsmen_season = top_batsmen_season.sort_values(['season', 'batsman_runs'], ascending=[True, False]).groupby('season').head(10)

# Plotting top batsmen each season
plt.figure(figsize=(15, 15))
sns.barplot(x='season', y='batsman_runs', hue='batsman', data=top_batsmen_season, palette='viridis')
plt.title('Top Batsmen Each Season')
plt.xlabel('Season')
plt.ylabel('Total Runs')
plt.xticks(rotation=45)
plt.legend(title='Batsmen', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
# Top bowlers each season
top_bowlers_season = deliveries[deliveries['dismissal_kind'].notnull() & (deliveries['dismissal_kind'] != 'run out')].groupby(['season', 'bowler'])['dismissal_kind'].count().reset_index()
top_bowlers_season = top_bowlers_season.sort_values(['season', 'dismissal_kind'], ascending=[True, False]).groupby('season').head(10)

# Plotting top bowlers each season
plt.figure(figsize=(15, 15))
sns.barplot(x='season', y='dismissal_kind', hue='bowler', data=top_bowlers_season, palette='viridis')
plt.title('Top Bowlers Each Season')
plt.xlabel('Season')
plt.ylabel('Total Wickets')
plt.xticks(rotation=45)
plt.legend(title='Bowlers', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
# Venue Analysis
```

```python
# Matches played at each venue
matches_per_venue = matches['venue'].value_counts()

# Plotting matches played at each venue
plt.figure(figsize=(15, 8))
sns.barplot(x=matches_per_venue.index, y=matches_per_venue.values, palette='viridis')
plt.title('Matches Played at Each Venue')
plt.xlabel('Venue')
plt.ylabel('Number of Matches')
plt.xticks(rotation=90)
plt.show()

```

```python
# Top 10 highest run scorers in an innings
top_10_scores = delivery.groupby(['match_id', 'batsman'])['batsman_runs'].sum().reset_index().sort_values(by='batsman_runs', ascending=False).head(10)

# Plotting top 10 highest run scorers in an innings
plt.figure(figsize=(15, 8))
sns.barplot(x='batsman', y='batsman_runs', data=top_10_scores, palette='viridis')
plt.title('Top 10 Highest Run Scorers in an Innings')
plt.xlabel('Batsman')
plt.ylabel('Runs')
plt.xticks(rotation=45)
plt.show()
```

```python
# Top 10 best bowling figures in an innings
best_bowling = valid_dismissals.groupby(['match_id', 'bowler'])['dismissal_kind'].count().reset_index().sort_values(by='dismissal_kind', ascending=False).head(10)

# Plotting top 10 best bowling figures in an innings
plt.figure(figsize=(15, 8))
sns.barplot(x='bowler', y='dismissal_kind', data=best_bowling, palette='viridis')
plt.title('Top 10 Best Bowling Figures in an Innings')
plt.xlabel('Bowler')
plt.ylabel('Wickets')
plt.xticks(rotation=45)
plt.show()
```

```python
# Identify the most successful teams
```

```python
# Step 1: Calculate the total matches played by each team
team1_matches = matches['team1'].value_counts()
team2_matches = matches['team2'].value_counts()
total_matches = team1_matches.add(team2_matches, fill_value=0).sort_values(ascending=False)
```

#### We count the number of occurrences of each team in both columns to determine how many matches each team has played.

```python

# Step 2: Calculate the total wins by each team
total_wins = matches['winner'].value_counts().sort_values(ascending=False)
```

#### We count the number of times each team appears in the 'winner' column of the 'matches' dataset to determine the total number of matches each team has won.

```python

# Step 3: Calculate the winning percentage
winning_percentage = (total_wins / total_matches) * 100

# Create a DataFrame to hold the results
team_performance = pd.DataFrame({
    'Total Matches': total_matches,
    'Total Wins': total_wins,
    'Winning Percentage': winning_percentage
}).sort_values(by='Winning Percentage', ascending=False).dropna()

# Display the DataFrame
print(team_performance)
```

#### We calculate the winning percentage for each team by dividing the total number of wins by the total number of matches played and then multiplying by 100 to get a percentage.

```python

# Step 4: Visualize the results

# Plot total wins
plt.figure(figsize=(15, 6))
sns.barplot(x=total_wins.index, y=total_wins.values, palette='viridis')
plt.title('Total Wins by Each Team')
plt.xlabel('Team')
plt.ylabel('Total Wins')
plt.xticks(rotation=90)
plt.show()
```

```python

# Plot winning percentage
plt.figure(figsize=(15, 6))
sns.barplot(x=winning_percentage.index, y=winning_percentage.values, palette='viridis')
plt.title('Winning Percentage by Each Team')
plt.xlabel('Team')
plt.ylabel('Winning Percentage')
plt.xticks(rotation=90)
plt.show()
```
